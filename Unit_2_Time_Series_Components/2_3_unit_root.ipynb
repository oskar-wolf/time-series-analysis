```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# %% set font size for graphs and seed
plt.rcParams.update({'font.size': 18})
np.random.seed(97)
```

```python
def ar_sim(alpha,sigma =1, x0=0,N=100):
    
    x = np.zeros(N)

    x[0] = x0

    for i in range(N-1):
        x[i+1] = alpha * x0 + sigma * np.random.randn()

        x0 = x[i+1]

    return(x)
```

```python
x_stat = ar_sim(alpha=0.8)
x_unit_root = ar_sim(alpha=1)
x_non_stat = ar_sim(alpha =1.1)
```

```python
fig, axes = plt.subplots(3, figsize=(16,18), dpi = 100)

plt. subplots_adjust(hspace = 0.3)

axes[0].plot(x_stat,color='k')
axes[0].set_title('Sequence: A')
axes[0].set(xlabel='Time')
axes[0].grid()

axes[1].plot(x_unit_root,color = 'k')
axes[1].set_title('Sequence: B')
axes[1].set(xlabel='Time')
axes[1].grid()

axes[2].plot(x_non_stat,color = 'k')
axes[2].set_title('Sequence: C')
axes[2].set(xlabel='Time')
axes[2].grid()
```
