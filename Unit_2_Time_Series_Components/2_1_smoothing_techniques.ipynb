```python
import pandas as pd

df = pd.read_csv(r'..\Unit_4_ARMA_Models\Global_Temperature.csv', \
    sep=';',index_col='Year')
```

```python
roll_temp_mean = df.rolling(window=5,center=True).mean()
```

```python
from statsmodels.nonparametric.kernel_regression import KernelReg
```

```python
import matplotlib.pyplot as plt 
```

```python
kr = KernelReg(df['Global_Temp'].values, \
    df.index, var_type='c',bw=[5])
```

```python
x_pred,x_std = kr.fit(df.index)
```

```python
plt.figure(figsize=(16, 5), dpi=100)

plt.plot(df.index,df.Global_Temp)

plt.plot(df.index, roll_temp_mean['Global_Temp'], color='green', label='5-Year Moving Average')

plt.plot(df.index,x_pred,':', \
    color='blue', label='&b = 5$')

plt.legend()
```
