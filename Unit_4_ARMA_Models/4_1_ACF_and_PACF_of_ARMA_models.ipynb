```python
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess 
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# set ar parameters
# for zero-lag, add 1
# for lag 1, e.g. AR(1) model, set phi to 0.8
# (given as negative parameter, see polynomial formula)

ar = np.array([1,-0.8])
```

```python
arma_model = ArmaProcess(ar=ar,ma=None)
```

```python
simulated_ar = arma_model.generate_sample(nsample=1000)
```

```python
plt.rcParams.update({'font.size' : 18})

fig, axs = plt.subplots(2,1, figsize = (16,10), dpi=100)
plt.subplots_adjust(hspace = 0.3)

plot_acf(simulated_ar,alpha=0.05,ax=axs[0])
axs[0].set_xlabel('Lag')
axs[0].set_ylabel('Correlation')
axs[0].set_title(r'ACF AR(1) Model with $\phi=0.8$')

plot_pacf(simulated_ar,alpha = 0.05,ax=axs[1])
axs[1].set_xlabel('Lag')
axs[1].set_ylabel('Correlation')
axs[1].set_title('PACF AR(1) Model with $\phi=0.8$')
```

```python
ma = np.array([1,-0.8])
arma_model = ArmaProcess(ar=None, ma=ma)
simulated_ar = \
    arma_model.generate_sample(nsample=10000)

# %% figure preparation
plt.rcParams.update({'font.size': 18})
fig, axes = plt.subplots(2, 1, figsize=(16,10), dpi=100)
plt.subplots_adjust(hspace=0.3)

# plot ACF
plot_acf(simulated_ar, alpha=0.05, ax=axes[0])
axes[0].set_xlabel('Lag')
axes[0].set_ylabel('Correlation')
axes[0].set_title(r'ACF MR(1) Model with $\phi=0.8$')

# plot PACF
plot_pacf(simulated_ar, alpha=0.05, ax=axes[1])
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('Correlation')
axes[1].set_title(r'PACF MR(1) Model with $\phi=0.8$')
```

```python
ar = np.array([1,-0.4])
ma = np.array([1,0.8])

arma_model = ArmaProcess(ar=ar,ma=ma)
simulated_arma = arma_model.generate_sample(nsample=1000)

plt.rcParams.update({'font.size': 18})

fig,axs = plt.subplots(2,1,figsize=(16,10), dpi = 100)

plt.subplots_adjust(hspace = 0.3)

plot_acf(simulated_arma, alpha=0.05, ax=axs[0])
axes[0].set_xlabel('Lag')
axes[0].set_ylabel('Correlation')
axes[0].set_title('ACF ARMA(1,1) Model $\phi=-0.4$ and $\theta=0.8$')

plot_pacf(simulated_arma, alpha=0.05,ax=axs[1])
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('Correlation')
axes[1].set_title('PACF ARMA(1,1) Model $\phi=-0.4$ and $\theta=0.8$')
```
